## What exactly is Gradient Descent?
Gradient Descent is an optimization algorithm used to minimize the loss function and update model parameters in machine learning. There are three main types:

1Ô∏è‚É£ **Batch Gradient Descent (BGD)**
üìå Updates parameters using the entire dataset at each step.

üîπ How it works:

Computes gradients over the entire dataset.

Updates weights only once per epoch (after all samples are processed).

‚úÖ Pros:

‚úî More stable updates, converges smoothly.

‚úî Suitable for small datasets.

‚ùå Cons:

‚ùå Slow for large datasets (expensive computations).

‚ùå Can get stuck in local minima.
******************************************************
2Ô∏è‚É£ **Stochastic Gradient Descent (SGD)**
üìå Updates parameters after every single training example (one at a time).

üîπ How it works:

Randomly selects one sample and updates weights immediately.

Faster but more noisy updates.

‚úÖ Pros:
‚úî Works well for large datasets.

‚úî Can escape local minima due to randomness.

‚ùå Cons:

‚ùå High variance in updates (can oscillate).

‚ùå Convergence is less stable.
 ********************************************************

3Ô∏è‚É£ **Mini-Batch Gradient Descent**
üìå Updates parameters using a small batch of training examples.

üîπ How it works:

A compromise between Batch GD and SGD.

Divides data into mini-batches (e.g., 32 or 64 samples per batch).
Computes gradients for each mini-batch and updates weights.

‚úÖ Pros:
‚úî Faster than Batch GD and less noisy than SGD.

‚úî Efficient for large datasets.

‚úî Takes advantage of vectorized operations (faster computation).

‚ùå Cons:

‚ùå Requires hyperparameter tuning (choosing the batch size).


‚úî **Batch GD: Best for small datasets, but slow for large ones.**

‚úî **SGD: Works well for big datasets, but updates can be unstable.**

‚úî **Mini-Batch GD: The best compromise between speed and stability.**

The below process is just for understanding how Gradient descent works step by step.

‚úî Gradient Descent iteratively updates b to minimize loss.

‚úî Regression lines shift step by step, showing how the model improves.

‚úî Learning rate controls how fast the updates happen.

```python
from sklearn.datasets import make_regression
import numpy as np
import pandas as pd
```

```python
X, y = make_regression(n_samples=4, n_features=1, n_targets=1, n_informative=1, noise=80)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
#Applying OLS
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X,y)
```

```python
model.coef_
#slope
```

```python
model.intercept_
#intercept
```

```python
plt.scatter(X,y)
plt.plot(X, model.predict(X), color='red')
```

Now, we will take random value of b, using Gradient decent, at the end will see that our value will be close to 33.83 as seen in OLS method.

```python
#Let's apply Gradient descent assuming the constant m is 135.72
#Let's assume the starting value for intercept b = 0

y_pred = ((135.72*X)+0).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X, model.predict(X), color='red', label='OLS')
plt.plot(X, y_pred, color='blue', label='b=0')
plt.legend()
plt.show()
```

We will be applying gradient descent technique to bring the blue line(here b'value is taken as 0) close to red line.

```python
m = 135.72
b = 0

loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
#Let's take learning rate as 0.1
lr = 0.1

step_size = loss_slope*lr
step_size
```

```python
#Calculating the new intercept
b = b - step_size
b
```

```python
y_pred1 = ((135.72*X)+b).reshape(4)

plt.scatter(X,y)
plt.plot(X, model.predict(X), color='black', label = 'OLS')
plt.plot(X, y_pred1, color='green', label = 'b={}'.format(b))
plt.plot(X, y_pred, color='red', label = 'b = 0')
plt.legend()
plt.show()
```

```python
#Iteration 2
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python

y_pred2 = ((135.72*X)+b).reshape(4)

plt.scatter(X,y)
plt.plot(X, model.predict(X), color='black', label = 'OLS')
plt.plot(X, y_pred2, color='yellow', label = 'b={}'.format(b))
plt.plot(X, y_pred1, color='green', label = 'b={}'.format(b))
plt.plot(X, y_pred, color='red', label = 'b = 0')
plt.legend()
plt.show()
```

```python
#3rd Iteration
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python

y_pred3 = ((135.72*X)+b).reshape(4)

plt.scatter(X,y)
plt.plot(X, model.predict(X), color='black', label = 'OLS')
plt.plot(X, y_pred3, color='grey', label = 'b={}'.format(b))
plt.plot(X, y_pred2, color='yellow', label = 'b={}'.format(b))
plt.plot(X, y_pred1, color='green', label = 'b={}'.format(b))
plt.plot(X, y_pred, color='red', label = 'b = 0')
plt.legend()
plt.show()
```

```python
#4th iteration
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred4 = ((135.72*X)+b).reshape(4)

plt.scatter(X,y)
plt.plot(X, model.predict(X), color='black', label = 'OLS')
plt.plot(X, y_pred4, color='orange', label = 'b={}'.format(b))
plt.plot(X, y_pred3, color='grey', label = 'b={}'.format(b))
plt.plot(X, y_pred2, color='yellow', label = 'b={}'.format(b))
plt.plot(X, y_pred1, color='green', label = 'b={}'.format(b))
plt.plot(X, y_pred, color='red', label = 'b = 0')
plt.legend()
plt.show()
```

* The b value is close 97.74 that is almost exactly equal to intercept found in OLS method.
* The black line and Orange line are almost exactly lie on top of each other.

#We have used for loop here and epochs set to 10. The above iteration process is just for understanding(breakdown). The below is the efficient way.

```python
#here b we have taken as -100 ( instead of 0 in above case)
b = -100
m = 135.72
lr = 0.1

epochs = 10

for i in range(epochs):
  loss_slope = -2 * np.sum(y - m*X.ravel() - b)
  b = b - (lr * loss_slope)

  y_pred = m * X + b
  plt.plot(X, y_pred)

plt.scatter(X,y)
```

- Computes the gradient (loss_slope) for each iteration.

- Updates b by subtracting the gradient multiplied by the learning rate.

- Plots updated regression lines for every iteration to visualize convergence.
